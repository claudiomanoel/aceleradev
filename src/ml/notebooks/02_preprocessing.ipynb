```python
#Preencher NAs
gc.collect()
boolean_types = df.columns[df.dtypes == 'bool'].tolist()
object_types = df.columns[df.dtypes == 'object'].tolist()
float_types = df.columns[df.dtypes == 'float'].tolist()
df[boolean_types] = df[boolean_types].fillna(value=0)
df[object_types].fillna(value='',inplace=True)
df.fillna({x:False for x in boolean_types}, inplace=True)
df.fillna({x:'' for x in object_types}, inplace=True) 
df.fillna({x:0 for x in float_types}, inplace=True) 
#Checar se existem valores nulos ainda na base
df.isnull().values.any()
```

```python
#Correção das colunas de flags não suportadas
df.loc[df.fl_spa == "", 'fl_spa'] = False
df.loc[df.fl_antt == "", 'fl_antt'] = False
df.loc[df.fl_veiculo == "", 'fl_veiculo'] = False
df.loc[df.fl_optante_simples == "", 'fl_optante_simples'] = False
df.loc[df.fl_optante_simei == "", 'fl_optante_simei'] = False
df.loc[df.fl_simples_irregular == "", 'fl_simples_irregular'] = False
df.loc[df.fl_passivel_iss == "", 'fl_passivel_iss'] = False

df['fl_spa'] = df['fl_spa'].astype('bool')
df['fl_antt'] = df['fl_antt'].astype('bool')
df['fl_veiculo'] = df['fl_veiculo'].astype('bool')
df['fl_optante_simples'] = df['fl_optante_simples'].astype('bool')
df['fl_simples_irregular'] = df['fl_simples_irregular'].astype('bool')
df['fl_passivel_iss'] = df['fl_passivel_iss'].astype('bool')
```

```python
#Processar OneHotEncoder para colunas com baixa cardinalidade
category_columns = ['sg_uf', 'natureza_juridica_macro', 'de_ramo', 'setor', 'idade_emp_cat', 'fl_rm', 'nm_segmento',
                'sg_uf_matriz', 'de_saude_tributaria', 'de_saude_rescencia', 'de_nivel_atividade', 'nm_meso_regiao',
                   'de_faixa_faturamento_estimado', 'fl_optante_simei']
df = pd.get_dummies(df, prefix=category_columns, columns=category_columns)
df.shape
```

```python
#Processar LabelEncoder para columnas com alta cardinalidade
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df['de_natureza_juridica'] = labelencoder.fit_transform(df['de_natureza_juridica'])
df['dt_situacao'] = labelencoder.fit_transform(df['dt_situacao'])
df['nm_divisao'] = labelencoder.fit_transform(df['nm_divisao'])
df['nm_micro_regiao'] = labelencoder.fit_transform(df['nm_micro_regiao'])
df['de_faixa_faturamento_estimado_grupo'] = labelencoder.fit_transform(df['de_faixa_faturamento_estimado_grupo'])
```

```python
#Padronizar as colunas
gc.collect()
from sklearn.preprocessing import StandardScaler
id_value = df['id']
df = df.drop(['id'], axis=1)
scaled_features = StandardScaler().fit_transform(df.values)
df = pd.DataFrame(scaled_features, index=df.index, columns=df.columns)
df
```

```python
#Redução de dimensionalidade
from sklearn.decomposition import PCA

pca = PCA(.95)
principal_components = pca.fit_transform(df)
print("Number of components:", pca.n_components_)
```

```python
#Salvamento dos Components principais
import pickle
with open('../../../output/principal_components.pickle', 'wb') as f:
    pickle.dump(principal_components, f)
```
